### **Task 1**
Напишите функцию fibonacci(index), которая находит число Фибоначчи с нужным номером. Аргумент функции index — целое положительное число.

Последовательность Фибоначчи задается формулой:

F(1) = 0

F(2) = 1

F(N) = F(N - 1) + F(N - 2) для N >= 3

То есть, каждое последующее число равно сумме двух предыдущих чисел. Если требуется больше информации о данной последовательности, вы можете обратиться, например, к источнику.

```python
def fibonacci(index):
  f1, f2 = 0, 1
  while index > 0:
    index -= 1
    if index > 0:
      f1, f2 = f1+f2, f1
    else:
      return f1
  return f1

i = int(input())
print(fibonacci(i))
```

### Task 2
Напишите функцию last_discharge(a), которая в качестве аргумента получает строку, содержащую число a, и возвращает это же число, уменьшенное на единицу в последнем его разряде в виде строки.

Обратите внимание:

●     в последнем разряде может быть ноль,

●     входная строка может представлять как целое число, так и число с плавающей точкой,

●     входная строка содержит число строго больше нуля.

```python
def last_discharge(a):
  a_list = list(map(str, a.split('.')))
  if len(a_list) > 1:
    if (int(a_list[-1]) == 0):
      minus = a_list[-1]
      minus = list(map(str,str(minus)))
      minus[-1] = '1'
      a_list = float('.'.join(a_list))
      minus[-1] = '1'
      minus = ''.join(minus)
      minus = '.'.join(['0', minus])
      a_list -= float(minus)
      a_list= str(a_list)
    else:
      a_list = list(map(int, a.split('.')))
      a_list[-1] = a_list[-1] -1
      a_list = list(map(str, a_list))
      a_list = '.'.join(a_list)
  else:
    a_list = str(int(a_list[-1]) - 1)
  return a_list

a = input()
last_discharge(a)
```

```python
with open(input(), 'r') as f:
    my_dict = {}
    for line in f:
        words = list(map(str, line.split()))
        for word in words:
            if word not in my_dict:
                my_dict[word] = 1
            else:
                my_dict[word] += 1

my_dict = sorted(my_dict.items(),  key=lambda x:x[0])
my_dict = list(sorted(my_dict,  key=lambda x:x[1], reverse = True))
for a, b in my_dict:
    print(b, a)
f.close()
```
