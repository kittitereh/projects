```python
@dataclass
class Msg:
    header: str
    text:   str
    spam:   bool
```

```python
@dataclass
class WordFreq:
    spam:  int = 0
    legit: int = 0
```

```python
def read_file(file_name) -> Msg:
    with open(file_name, "r") as file:
        full_text = file.readlines()
        first_line = full_text[0]
        
        body_lines = full_text[2:]
        
        body = [word for line in body_lines for word in line.split()]
        
        msg_header = first_line.split()[1:]
        
        spam_flag = "spm" in file_name
        
        return Msg(header = msg_header, text = body, spam = "spm" in file_name)
```

```python
def fill_counters_from_msg(msg: Msg, words_freqs):
    def fill_counters_from(sq: Sequence):
#         print(sq)
        for word in sq:
            if word not in words_freqs:
                words_freqs[word] = WordFreq()
#             else:
#                 print(word)
        
            word_freq = words_freqs[word]
        
            if msg.spam:
                word_freq.spam  += 1
            else:
                word_freq.legit += 1
    
    fill_counters_from(msg.header)    
    fill_counters_from(msg.text)
```
